  - ‡∏™‡∏≥‡∏´‡∏£‡∏±‡∏ö‡πÅ‡∏ï‡πà‡∏•‡∏∞ dimension ‡πÉ‡∏´‡πâ‡∏™‡∏£‡πâ‡∏≤‡∏á ‡∏ï‡∏±‡∏ß‡∏≠‡∏¢‡πà‡∏≤‡∏á‡∏Ñ‡πà‡∏≤‡∏Ç‡∏≠‡∏á dimension ‡∏ô‡∏±‡πâ‡∏ô‡πÜ ‡∏≠‡∏¢‡πà‡∏≤‡∏á‡∏ô‡πâ‡∏≠‡∏¢ ‡∏™‡∏≤‡∏°‡∏Ñ‡πà‡∏≤.

  - ‡πÄ‡∏Ç‡∏µ‡∏¢‡∏ô prompt ‡∏™‡∏≥‡∏´‡∏£‡∏±‡∏ö LLM ‡∏ó‡∏µ‡πà‡∏à‡∏∞‡∏£‡∏±‡∏ö 10 tuples (‡πÄ‡∏•‡∏∑‡∏≠‡∏Å‡πÄ‡∏≠‡∏á) ‡πÄ‡∏û‡∏∑‡πà‡∏≠‡∏™‡∏£‡πâ‡∏≤‡∏á query ‡∏à‡∏£‡∏¥‡∏á‡∏≠‡∏≠‡∏Å‡∏°‡∏≤‡πÉ‡∏ä‡πâ‡∏ó‡∏î‡∏™‡∏≠‡∏ö ‡πÉ‡∏´‡πâ‡∏™‡∏£‡πâ‡∏≤‡∏á 3 queries ‡∏ï‡πà‡∏≠ 1 tuple. ‡∏£‡∏ß‡∏°‡∏ó‡∏±‡πâ‡∏á‡∏´‡∏°‡∏î 30 ‡∏≠‡∏±‡∏ô
  - ‡∏´‡∏•‡∏±‡∏á‡∏à‡∏≤‡∏Å‡∏ô‡∏±‡πâ‡∏ô‡πÉ‡∏´‡πâ‡πÄ‡∏ä‡πá‡∏Ñ‡∏ß‡πà‡∏≤ ‡∏ó‡∏µ‡πà‡∏™‡∏£‡πâ‡∏≤‡∏á‡∏≠‡∏≠‡∏Å‡∏°‡∏≤‡∏ô‡∏±‡πâ‡∏ô ‡∏≠‡∏±‡∏ô‡∏ô‡∏±‡πâ‡∏ô‡∏™‡∏°‡∏à‡∏£‡∏¥‡∏á‡πÅ‡∏•‡∏∞‡∏Ñ‡∏¥‡∏î‡∏ß‡πà‡∏≤‡πÄ‡∏õ‡πá‡∏ô‡∏Ñ‡∏≥‡∏ñ‡∏≤‡∏°‡∏ó‡∏µ‡πà‡πÄ‡∏õ‡πá‡∏ô‡πÑ‡∏õ‡πÑ‡∏î‡πâ‡∏ö‡πâ‡∏≤‡∏á ‡πÅ‡∏•‡∏∞‡πÄ‡∏•‡∏∑‡∏≠‡∏Å‡∏°‡∏≤‡πÉ‡∏™‡πà‡πÉ‡∏ô dataset ‡πÑ‡∏ß‡πâ‡∏ó‡∏î‡∏™‡∏≠‡∏ö
  

**‡∏Ñ‡∏≥‡πÅ‡∏ô‡∏∞‡∏ô‡∏≥**: ‡∏î‡∏π‡∏ï‡∏±‡∏ß‡∏≠‡∏¢‡πà‡∏≤‡∏á‡∏à‡∏≤‡∏Å slide ‡πÑ‡∏î‡πâ

**Answer**: ‡∏ï‡∏≠‡∏ö Key Dimensions ‡∏û‡∏£‡πâ‡∏≠‡∏°‡∏Ñ‡∏≥‡∏≠‡∏ò‡∏¥‡∏ö‡∏≤‡∏¢ ‡∏î‡πâ‡∏≤‡∏ô‡∏•‡πà‡∏≤‡∏á
1. ‡∏õ‡∏£‡∏∞‡πÄ‡∏†‡∏ó‡∏≠‡∏≤‡∏´‡∏≤‡∏£
2. ‡∏†‡∏π‡∏°‡∏¥‡∏†‡∏≤‡∏Ñ (‡πÄ‡∏ä‡πà‡∏ô‡∏≠‡∏≤‡∏´‡∏≤‡∏£‡πÄ‡∏´‡∏ô‡∏∑‡∏≠, ‡∏Å‡∏•‡∏≤‡∏á, ‡∏≠‡∏µ‡∏™‡∏≤‡∏ô, ‡πÉ‡∏ï‡πâ)
3. ‡∏Ç‡πâ‡∏≠‡∏à‡∏≥‡∏Å‡∏±‡∏î (‡πÄ‡∏ä‡πà‡∏ô ‡πÑ‡∏°‡πà‡∏Å‡∏¥‡∏ô‡πÄ‡∏ú‡πá‡∏î, ‡πÄ‡∏à, ‡∏≠‡∏≤‡∏´‡∏≤‡∏£‡∏Æ‡∏≤‡∏•‡∏≤‡∏•)

````python
import json
tuple_gen_prompt = """‡∏Å‡∏£‡∏∏‡∏ì‡∏≤‡∏™‡∏£‡πâ‡∏≤‡∏á 15 combinations ‡∏ó‡∏µ‡πà‡πÑ‡∏°‡πà‡∏ã‡πâ‡∏≥‡∏Å‡∏±‡∏ô‡∏Ç‡∏≠‡∏á‡∏°‡∏¥‡∏ï‡∏¥‡∏ï‡πà‡∏≠‡πÑ‡∏õ‡∏ô‡∏µ‡πâ‡∏™‡∏≥‡∏´‡∏£‡∏±‡∏ö‡∏™‡∏π‡∏ï‡∏£‡∏≠‡∏≤‡∏´‡∏≤‡∏£‡πÑ‡∏ó‡∏¢:

‡∏°‡∏¥‡∏ï‡∏¥‡∏ó‡∏µ‡πà 1 - ‡∏õ‡∏£‡∏∞‡πÄ‡∏†‡∏ó‡∏≠‡∏≤‡∏´‡∏≤‡∏£: ‡∏≠‡∏≤‡∏´‡∏≤‡∏£‡∏à‡∏≤‡∏ô‡∏´‡∏•‡∏±‡∏Å, ‡∏Ç‡∏≠‡∏á‡∏´‡∏ß‡∏≤‡∏ô, ‡∏≠‡∏≤‡∏´‡∏≤‡∏£‡πÄ‡∏ä‡πâ‡∏≤, ‡∏≠‡∏≤‡∏´‡∏≤‡∏£‡∏ß‡πà‡∏≤‡∏á, ‡πÄ‡∏Ñ‡∏£‡∏∑‡πà‡∏≠‡∏á‡∏î‡∏∑‡πà‡∏°
‡∏°‡∏¥‡∏ï‡∏¥‡∏ó‡∏µ‡πà 2 - ‡∏†‡∏π‡∏°‡∏¥‡∏†‡∏≤‡∏Ñ: ‡∏†‡∏≤‡∏Ñ‡∏Å‡∏•‡∏≤‡∏á, ‡∏†‡∏≤‡∏Ñ‡πÄ‡∏´‡∏ô‡∏∑‡∏≠, ‡∏†‡∏≤‡∏Ñ‡∏≠‡∏µ‡∏™‡∏≤‡∏ô, ‡∏†‡∏≤‡∏Ñ‡πÉ‡∏ï‡πâ
‡∏°‡∏¥‡∏ï‡∏¥‡∏ó‡∏µ‡πà 3 - ‡∏Ç‡πâ‡∏≠‡∏à‡∏≥‡∏Å‡∏±‡∏î‡∏î‡πâ‡∏≤‡∏ô‡∏≠‡∏≤‡∏´‡∏≤‡∏£: ‡πÑ‡∏°‡πà‡∏°‡∏µ‡∏Ç‡πâ‡∏≠‡∏à‡∏≥‡∏Å‡∏±‡∏î, ‡πÄ‡∏à, ‡∏°‡∏±‡∏á‡∏™‡∏ß‡∏¥‡∏£‡∏±‡∏ï‡∏¥, ‡πÑ‡∏°‡πà‡∏ó‡∏≤‡∏ô‡πÄ‡∏ú‡πá‡∏î, ‡πÑ‡∏°‡πà‡∏ó‡∏≤‡∏ô‡∏≠‡∏≤‡∏´‡∏≤‡∏£‡∏ó‡∏∞‡πÄ‡∏•

‡πÉ‡∏´‡πâ output ‡πÄ‡∏õ‡πá‡∏ô JSON array ‡∏Ç‡∏≠‡∏á objects ‡∏ó‡∏µ‡πà‡∏°‡∏µ keys: meal_type, region, dietary_restriction

‡∏ï‡∏±‡∏ß‡∏≠‡∏¢‡πà‡∏≤‡∏á:
[
  {"meal_type": "‡∏≠‡∏≤‡∏´‡∏≤‡∏£‡∏à‡∏≤‡∏ô‡∏´‡∏•‡∏±‡∏Å", "region": "‡∏†‡∏≤‡∏Ñ‡∏Å‡∏•‡∏≤‡∏á", "dietary_restriction": "‡πÑ‡∏°‡πà‡∏°‡∏µ‡∏Ç‡πâ‡∏≠‡∏à‡∏≥‡∏Å‡∏±‡∏î"},
  {"meal_type": "‡∏Ç‡∏≠‡∏á‡∏´‡∏ß‡∏≤‡∏ô", "region": "‡∏†‡∏≤‡∏Ñ‡πÄ‡∏´‡∏ô‡∏∑‡∏≠", "dietary_restriction": "‡πÄ‡∏à"}
]

‡∏Å‡∏£‡∏∏‡∏ì‡∏≤‡∏™‡∏£‡πâ‡∏≤‡∏á 15 combinations ‡∏ó‡∏µ‡πà‡∏´‡∏•‡∏≤‡∏Å‡∏´‡∏•‡∏≤‡∏¢‡πÅ‡∏•‡∏∞‡πÑ‡∏°‡πà‡∏ã‡πâ‡∏≥‡∏Å‡∏±‡∏ô ‡∏ï‡∏≠‡∏ö‡πÄ‡∏â‡∏û‡∏≤‡∏∞ JSON array ‡πÄ‡∏ó‡πà‡∏≤‡∏ô‡∏±‡πâ‡∏ô ‡πÑ‡∏°‡πà‡∏ï‡πâ‡∏≠‡∏á‡∏°‡∏µ‡∏Ñ‡∏≥‡∏≠‡∏ò‡∏¥‡∏ö‡∏≤‡∏¢‡∏≠‡∏∑‡πà‡∏ô"""

# ‡πÄ‡∏ï‡∏¥‡∏° code ‡πÄ‡∏û‡∏∑‡πà‡∏≠ gen tuple ‡∏≠‡∏≠‡∏Å‡∏°‡∏≤ 15 ‡∏≠‡∏±‡∏ô‡∏ó‡∏µ‡πà‡πÑ‡∏°‡πà‡∏ã‡πâ‡∏≥‡∏Å‡∏±‡∏ô
unique_tuples = tuple()
try:
    tuples_response = generate("You are a helpful assistant that generates JSON data. Return only valid JSON without any markdown formatting.", tuple_gen_prompt)
    # ‡∏•‡∏ö markdown code blocks ‡∏ñ‡πâ‡∏≤‡∏°‡∏µ
    tuples_clean = tuples_response.replace("```json", "").replace("```", "").strip()
    unique_tuples = json.loads(tuples_clean)
    print("‚úì Generated tuples from LLM successfully")
except:
    print("‚ö† Using predefined tuples (LLM not available or parsing failed)")

# uniques_tuple
print(unique_tuples)
````

````python
import random
query_prompt = """‡∏à‡∏≤‡∏Å combination ‡∏ô‡∏µ‡πâ:
- ‡∏õ‡∏£‡∏∞‡πÄ‡∏†‡∏ó‡∏≠‡∏≤‡∏´‡∏≤‡∏£: {meal_type}
- ‡∏†‡∏π‡∏°‡∏¥‡∏†‡∏≤‡∏Ñ: {region}
- ‡∏Ç‡πâ‡∏≠‡∏à‡∏≥‡∏Å‡∏±‡∏î‡∏î‡πâ‡∏≤‡∏ô‡∏≠‡∏≤‡∏´‡∏≤‡∏£: {dietary_restriction}

‡∏Å‡∏£‡∏∏‡∏ì‡∏≤‡∏™‡∏£‡πâ‡∏≤‡∏á 3 ‡∏Ñ‡∏≥‡∏ñ‡∏≤‡∏°‡∏ó‡∏µ‡πà‡∏ú‡∏π‡πâ‡πÉ‡∏ä‡πâ‡∏à‡∏£‡∏¥‡∏á‡∏≠‡∏≤‡∏à‡∏à‡∏∞‡∏ñ‡∏≤‡∏°‡∏°‡∏≤‡πÉ‡∏ô‡∏£‡∏π‡∏õ‡πÅ‡∏ö‡∏ö‡∏†‡∏≤‡∏©‡∏≤‡∏ò‡∏£‡∏£‡∏°‡∏ä‡∏≤‡∏ï‡∏¥‡∏ó‡∏µ‡πà‡∏´‡∏•‡∏≤‡∏Å‡∏´‡∏•‡∏≤‡∏¢ ‡∏ã‡∏∂‡πà‡∏á‡∏™‡∏≠‡∏î‡∏Ñ‡∏•‡πâ‡∏≠‡∏á‡∏Å‡∏±‡∏ö combination ‡∏ô‡∏µ‡πâ

‡∏Ñ‡∏≥‡∏ñ‡∏≤‡∏°‡∏Ñ‡∏ß‡∏£‡∏à‡∏∞:
- ‡πÄ‡∏õ‡πá‡∏ô‡∏†‡∏≤‡∏©‡∏≤‡πÑ‡∏ó‡∏¢‡∏ó‡∏µ‡πà‡πÄ‡∏õ‡πá‡∏ô‡∏ò‡∏£‡∏£‡∏°‡∏ä‡∏≤‡∏ï‡∏¥‡πÅ‡∏•‡∏∞‡πÉ‡∏ä‡πâ‡πÉ‡∏ô‡∏ä‡∏µ‡∏ß‡∏¥‡∏ï‡∏õ‡∏£‡∏∞‡∏à‡∏≥‡∏ß‡∏±‡∏ô
- ‡∏°‡∏µ‡∏Ñ‡∏ß‡∏≤‡∏°‡∏´‡∏•‡∏≤‡∏Å‡∏´‡∏•‡∏≤‡∏¢‡πÉ‡∏ô‡∏£‡∏π‡∏õ‡πÅ‡∏ö‡∏ö‡∏Å‡∏≤‡∏£‡∏ñ‡∏≤‡∏° (‡∏ï‡∏£‡∏á‡πÑ‡∏õ‡∏ï‡∏£‡∏á‡∏°‡∏≤, ‡∏≠‡πâ‡∏≠‡∏°, ‡∏™‡∏∏‡∏†‡∏≤‡∏û, ‡∏Å‡∏±‡∏ô‡πÄ‡∏≠‡∏á)
- ‡∏™‡∏∞‡∏ó‡πâ‡∏≠‡∏ô‡∏ñ‡∏∂‡∏á‡∏ö‡∏£‡∏¥‡∏ö‡∏ó‡πÅ‡∏•‡∏∞‡∏Ñ‡∏ß‡∏≤‡∏°‡∏ï‡πâ‡∏≠‡∏á‡∏Å‡∏≤‡∏£‡∏à‡∏£‡∏¥‡∏á‡∏Ç‡∏≠‡∏á‡∏ú‡∏π‡πâ‡πÉ‡∏ä‡πâ

‡πÉ‡∏´‡πâ output ‡πÄ‡∏õ‡πá‡∏ô JSON array ‡∏Ç‡∏≠‡∏á strings ‡πÄ‡∏ó‡πà‡∏≤‡∏ô‡∏±‡πâ‡∏ô ‡πÑ‡∏°‡πà‡∏ï‡πâ‡∏≠‡∏á‡∏°‡∏µ‡∏Ñ‡∏≥‡∏≠‡∏ò‡∏¥‡∏ö‡∏≤‡∏¢‡∏≠‡∏∑‡πà‡∏ô

‡∏ï‡∏±‡∏ß‡∏≠‡∏¢‡πà‡∏≤‡∏á:
["‡∏≠‡∏¢‡∏≤‡∏Å‡∏ó‡∏≥‡∏≠‡∏≤‡∏´‡∏≤‡∏£‡∏†‡∏≤‡∏Ñ‡πÄ‡∏´‡∏ô‡∏∑‡∏≠‡∏ó‡∏µ‡πà‡πÄ‡∏õ‡πá‡∏ô‡πÄ‡∏à‡∏™‡∏±‡∏Å‡∏≠‡∏¢‡πà‡∏≤‡∏á ‡∏°‡∏µ‡∏≠‡∏∞‡πÑ‡∏£‡πÅ‡∏ô‡∏∞‡∏ô‡∏≥‡∏ö‡πâ‡∏≤‡∏á", "‡∏°‡∏µ‡∏™‡∏π‡∏ï‡∏£‡∏≠‡∏≤‡∏´‡∏≤‡∏£‡πÄ‡∏ä‡πâ‡∏≤‡πÅ‡∏ö‡∏ö‡∏†‡∏≤‡∏Ñ‡∏Å‡∏•‡∏≤‡∏á‡∏ó‡∏µ‡πà‡πÑ‡∏°‡πà‡∏°‡∏µ‡∏≠‡∏≤‡∏´‡∏≤‡∏£‡∏ó‡∏∞‡πÄ‡∏•‡πÑ‡∏´‡∏°", "‡∏´‡∏¥‡∏ß‡∏Ç‡∏≠‡∏á‡∏´‡∏ß‡∏≤‡∏ô‡∏à‡∏±‡∏á ‡∏≠‡∏¢‡∏≤‡∏Å‡πÑ‡∏î‡πâ‡∏™‡∏π‡∏ï‡∏£‡∏Ç‡∏≠‡∏á‡∏´‡∏ß‡∏≤‡∏ô‡∏†‡∏≤‡∏Ñ‡πÉ‡∏ï‡πâ‡∏ó‡∏µ‡πà‡∏ó‡∏≥‡∏á‡πà‡∏≤‡∏¢‡πÜ"]"""
queries = []

sample_size = min(10, len(unique_tuples))
if sample_size < 10:
    print(f"‚ö† Warning: Only {sample_size} tuples available, generating queries for all of them")

selected_tuples = random.sample(unique_tuples, sample_size)

for idx, tuple_item in enumerate(selected_tuples, 1):
    try:
        formatted_prompt = query_prompt.format(
            meal_type=tuple_item["meal_type"],
            region=tuple_item["region"],
            dietary_restriction=tuple_item["dietary_restriction"]
        )

        queries_response = generate("You are a helpful assistant. Return only valid JSON.", formatted_prompt)
        queries_clean = queries_response.replace("```json", "").replace("```", "").strip()
        tuple_queries = json.loads(queries_clean)

        # ‡∏ï‡∏£‡∏ß‡∏à‡∏™‡∏≠‡∏ö‡∏ß‡πà‡∏≤‡πÄ‡∏õ‡πá‡∏ô list ‡πÅ‡∏•‡∏∞‡∏°‡∏µ‡∏Ç‡πâ‡∏≠‡∏°‡∏π‡∏•
        if isinstance(tuple_queries, list) and len(tuple_queries) > 0:
            queries.extend(tuple_queries)
            print(f"‚úì Generated {len(tuple_queries)} queries for combination {idx}/{sample_size}")
        else:
            print(f"‚ö† No queries generated for combination {idx}/{sample_size}")

    except json.JSONDecodeError as e:
        print(f"‚úó JSON parse error for combination {idx}/{sample_size}: {e}")
        print(f"  Response: {queries_clean[:100]}...")
    except KeyError as e:
        print(f"‚úó Missing key in tuple {idx}: {e}")
    except Exception as e:
        print(f"‚úó Error for combination {idx}/{sample_size}: {e}")


print(queries)
````

````python
dataset = []   # ‡∏ô‡∏≥ query ‡∏ó‡∏µ‡πà‡∏ú‡πà‡∏≤‡∏ô‡∏Å‡∏≤‡∏£‡∏Ñ‡∏±‡∏î‡∏Å‡∏£‡∏≠‡∏á‡∏°‡∏≤‡πÉ‡∏™‡πà
# ‡πÄ‡∏Å‡πá‡∏ö queries ‡∏ó‡∏µ‡πà‡πÑ‡∏î‡πâ‡∏à‡∏≤‡∏Å‡∏Å‡∏≤‡∏£ generate ‡∏•‡∏á‡πÉ‡∏ô dataset

for idx, query in enumerate(queries):
    # ‡∏´‡∏≤ tuple ‡∏ó‡∏µ‡πà‡∏ï‡∏£‡∏á‡∏Å‡∏±‡∏ö query ‡∏ô‡∏µ‡πâ (‡∏ó‡∏∏‡∏Å 3 queries ‡∏°‡∏≤‡∏à‡∏≤‡∏Å 1 tuple)
    tuple_idx = idx // 3
    if tuple_idx < len(selected_tuples):
        tuple_item = selected_tuples[tuple_idx]
        dataset.append({
            "query": query,
            "meal_type": tuple_item["meal_type"],
            "region": tuple_item["region"],
            "dietary_restriction": tuple_item["dietary_restriction"]
        })

print(f"‚úì Dataset created with {len(dataset)} queries")
print(f"\nSample from dataset:")
for i in range(min(3, len(dataset))):
    print(f"{i+1}. {dataset[i]['query']}")
    print(f"   - Type: {dataset[i]['meal_type']}, Region: {dataset[i]['region']}, Restriction: {dataset[i]['dietary_restriction']}\n")
````

````python
# Part 4.1: Run Bot on Synthetic Queries
results = []

print("Running Recipe Bot on all queries...")
print("="*80)

for idx, item in enumerate(dataset, 1):
    try:
        query = item["query"]
        print(f"\nProcessing query {idx}/{len(dataset)}...")
        print(f"Query: {query[:50]}...")

        # ‡πÄ‡∏£‡∏µ‡∏¢‡∏Å bot
        response = generate(recipe_bot_prompt, query)

        # ‡πÄ‡∏Å‡πá‡∏ö‡∏ú‡∏•‡∏•‡∏±‡∏û‡∏ò‡πå
        result = {
            "id": idx,
            "query": query,
            "response": response,
            "meal_type": item["meal_type"],
            "region": item["region"],
            "dietary_restriction": item["dietary_restriction"]
        }
        results.append(result)
        print(f"‚úì Success (response length: {len(response)} chars)")

    except Exception as e:
        print(f"‚úó Error: {e}")
        results.append({
            "id": idx,
            "query": item["query"],
            "response": f"ERROR: {str(e)}",
            "meal_type": item["meal_type"],
            "region": item["region"],
            "dietary_restriction": item["dietary_restriction"]
        })

print(f"\n{'='*80}")
print(f"Completed! Total results: {len(results)}")
print(f"{'='*80}")
````

````python
import csv

csv_data = []
for result in results:
    csv_data.append({
        'ID': result['id'],
        'Query': result['query'],
        'Meal_Type': result['meal_type'],
        'Region': result['region'],
        'Dietary_Restriction': result['dietary_restriction'],
        'Response_Preview': result['response'][:200] + '...' if len(result['response']) > 200 else result['response'],
        'Response_Length': len(result['response']),
        'Has_Error': 'Yes' if result['response'].startswith('ERROR') else 'No'
    })

with open('open_coding_analysis.csv', 'w', encoding='utf-8-sig', newline='') as f:
    if csv_data:
        writer = csv.DictWriter(f, fieldnames=csv_data[0].keys())
        writer.writeheader()
        writer.writerows(csv_data)

print("‚úì CSV file created: open_coding_analysis.csv")
print(f"Total rows: {len(csv_data)}")
print("\nColumns:")
for col in csv_data[0].keys():
    print(f"  - {col}")
````
